```python
import numpy as np
import matplotlib.pyplot as plt

D = np.array([407, 222, 170, 120, 101, 85.4, 68.5, 51.2, 34.1, 17.2, 8.59])
rho = np.array([19.8, 20.8, 20.0, 22.1, 23.5, 27.9, 30.7, 32.2, 50.4, 70.5, 126])

fig = plt.figure(figsize=(6,6))
ax1 = fig.add_subplot(211)
ax1.plot(1/D, rho, 'ro')
ax1.set_xlabel('1/D')
ax1.set_ylabel(r"$\rho$")
ax1.grid(True)

ax2 = fig.add_subplot(212)
ax2.plot(D, rho, 'ro')
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlabel('D')
ax2.set_ylabel(r"$\rho$")
ax2.grid(True)

plt.tight_layout()
plt.show()
```
