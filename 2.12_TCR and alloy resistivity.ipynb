```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def func(x, a, b):
  return a * x + b

wt = np.array([0, 2, 6, 11, 20])
rho = np.array([17, 50, 100, 150, 300])
tcr = np.array([4270, 1350, 550, 430, 160])

popt, pcov = curve_fit(func, 1/rho, tcr)

fig = plt.figure(figsize=(6,4))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(1/rho, tcr, 'ro')
ax1.plot(1/rho, func(1/rho, *popt), 'b-')
ax1.set_xlabel(r'1/$\rho$')
ax1.set_ylabel('TCR')
ax1.grid(True)

plt.show()
```
