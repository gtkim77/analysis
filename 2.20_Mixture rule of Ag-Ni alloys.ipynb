```python
import numpy as np
import matplotlib.pyplot as plt

def rho_eff(w, d, r):
  rho_Ag = 16.9 * 10**(-9)
  d_Ni = 8.91
  x = 71.0/16.9
  return rho_Ag/2*(3/(1+2*w*d/d_Ni*(1-x)/(1+2*x)) - 1)


ratio = np.array([0, 10, 15, 20, 30, 40, 100])
density = np.array([10.49, 10.25, 10.15, 10.05, 9.8, 9.7, 8.91])
rho = np.array([16.9, 18.7, 19.0, 20.0, 24.4, 27.0, 71.0])
resistivity_eff = rho_eff(ratio/100, density, rho) * 10**(9)

fig = plt.figure(figsize=(6,4))
ax1 = fig.add_subplot(111)
ax1.scatter(ratio, rho, color='r', label='experimental')
ax1.scatter(ratio, resistivity_eff, color='b', label='calculated by mixture rule')
ax1.set_xlabel('Ratio (weight % of Ni)')
ax1.set_ylabel('Resistivity (n$\Omega$m)')
ax1.grid(True)

plt.legend()
plt.show()
```
