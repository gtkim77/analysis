```python
import matplotlib.pyplot as plt
import numpy as np

Vg = np.array([44, 46, 48, 50, 52, 53.8, 56.2, 58.2, 60.4])
Ie = np.array([9.40, 20.4, 34.1, 61, 93.8, 142.5, 202, 279, 367]) * 10**-6

fig = plt.figure(figsize=(6,6))
ax1 = fig.add_subplot(2,1,1)
ax1.plot(Vg, Ie * 10**6, 'ro')
ax1.set_xlabel('V$_G$ (V)')
ax1.set_ylabel('I ($\mu$A)')
ax1.grid(True)

ax2 = fig.add_subplot(2,1,2)
ax2.plot(1/Vg, Ie/Vg**2, 'ro')
ax2.set_xlabel('1/V$_G$')
ax2.set_ylabel('ln(I$_A$/V$_G^2$)')
ax2.set_yscale('log')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def func(x, a, b):
    return a * x + b

VG = np.array([44,46,48,50,52,53.8,56.2,58.2,60.4])
Is = np.array([9.40,20.4,34.1,61,93.8,142.5,202,279,367])

x = 1/VG
y = np.log(Is / VG**2)

popt, perr = curve_fit(func, x, y)

print(popt)
plt.plot(x, y, 'ko')
plt.plot(x, func(x, *popt), 'b-')
plt.grid()
plt.show()
```
