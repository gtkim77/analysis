```python
import numpy as np

JTh = 3 * 10**4 * 1073**2 * np.exp(-2.6 * 1.6 * 10**(-19) / 1.38 / 1073 / 10**(-23))
Jox = 100 * 1073**2 * np.exp(-1 * 1.6 * 10**(-19) / 1.38 / 1073 / 10**(-23))

ATh = 10 / JTh
Aox = 10 / Jox

print('JTh: %.3e ATh: %.3e' % (JTh, ATh))
print('Jox: %.3e Aox: %.3e' % (Jox, Aox))
```

```python
import matplotlib.pyplot as plt

T = np.linspace(1000,2000,100)
J = 3 * 10**4 * T**2 * np.exp(-2.6 * 10**(-19) * 1.6 / 1.38 / 10**(-23) / T) - 2336

plt.plot(T,J)
plt.xlim([1700,1750])
plt.grid()
plt.show()
```

```python
from scipy.optimize import fsolve
f = lambda x: 3 * 10**4 * x**2 * np.exp(-2.6 * 10**(-19) * 1.6 / 1.38 / 10**(-23) / x) - 2336
fsolve(f, [1700, 1750])
```
