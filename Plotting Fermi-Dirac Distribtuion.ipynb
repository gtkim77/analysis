```python
import numpy as np
import matplotlib.pyplot as plt

e = 1.6 * 10**(-19)
T = 300
k = 1.38 * 10**(-23)
EF = 1.0

Ts = [5, 30, 300,1000]
E = np.linspace(0, 2, 100)

for T in Ts:
  plt.plot(E, 1 / (1 + np.exp((E-EF)*e/k/T)), label='T=%.1f K' % T)

plt.legend()
plt.grid()
plt.show()
```

```python
import scipy.integrate as integrate
import matplotlib.pyplot as plt
import numpy as np
h = 6.626 * 10**(-34)
k = 1.38 * 10**(-23)
c = 3.0 * 10**(8)
e = 1.6 * 10**(-19)

def radiation(x, T):
  return 2 * np.pi * (x * e)**3 / (c**2 * h**2) / (np.exp(x / (k * T / e) )) * e / h

Ts = [300, 1000, 3000, 5000, 10000]
Emax = 6
xs = np.linspace(0, Emax, 200)
for T in Ts:
    IT, error = integrate.quad(lambda x: radiation(x, T), 0, Emax)
    ys = radiation(xs, T) / IT
    index = np.argmax(ys)
    print('T = %.1f lambda_max = %.1f lambda_max * T = %.1e' % (T, h * c / (e * xs[index]) / 10**(-9), h * c / (e * xs[index]) * T))
    plt.plot(xs, ys, label='T=%.1f' % T)

plt.legend()
plt.show()
#result, error = integrate.quad(lambda x: radiation(x, 300), 0, 6)
#print(result)
```

```python
import numpy as np
import matplotlib.pyplot as plt
h = 6.626 * 10**(-34)
hbar = h / (2 * np.pi)
m = 9.1 * 10**(-31)
a = 4 * 10**(-9)
e = 1.6 * 10**(-19)
V0 = 0.4 * e


x = np.linspace(0, 10, 100)
yr = x * np.tan(x)
yl = (2 * m * a**2 * V0 / hbar**2 - x**2)**0.5
plt.plot(x, yr)
plt.plot(x, yl)

plt.ylim([0,30])

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
import scipy.integrate as integrate

def func(x, a3, a2, a1, a0):
  return a3 * x**3 + a2 * x**2 + a1 * x + a0

Ts = np.array([273, 300, 350, 400, 450, 500, 600, 700, 800, 900, 1000, 1100, 1200, 1300, 1400, 1600])
Ss = np.array([-4.04, -4.92, -6.33, -7.53, -8.59, -9.53, -11.22, -12.71, -14.14, -15.66, -17.21, -18.77, -20.29, -21.78, -23.18, -25.67])

popt, pcov = curve_fit(func, Ts, Ss)

print(integrate.quad(lambda x: func(x, *popt), 273, 473))

plt.plot(Ts, Ss, 'ko')
plt.plot(Ts, func(Ts, *popt), 'b-')
plt.grid()
plt.show()
```
