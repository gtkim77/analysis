```python
# https://realpython.com/linear-regression-in-python/

import numpy as np
import matplotlib.pyplot as plt

Metal = {}
Metal['Li'] = {'Mat': 6.94, 'Cs': 3.58}
Metal['Na'] = {'Mat': 22.99, 'Cs': 1.228}
Metal['Mg'] = {'Mat': 24.3, 'Cs': 1.023}
Metal['Al'] = {'Mat': 26.98, 'Cs': 0.897}
Metal['K'] = {'Mat': 39.1, 'Cs': 0.757}
Metal['Ca'] = {'Mat': 40.08, 'Cs': 0.647}
Metal['Ti'] = {'Mat': 47.87, 'Cs': 0.523}
Metal['V'] = {'Mat': 50.94, 'Cs': 0.489}
Metal['Cr'] = {'Mat': 51.99, 'Cs': 0.449}
Metal['Fe'] = {'Mat': 55.85, 'Cs': 0.444}
Metal['Co'] = {'Mat': 58.93, 'Cs': 0.421}
Metal['Cu'] = {'Mat': 58.93, 'Cs': 0.385}
Metal['Zn'] = {'Mat': 65.39, 'Cs': 0.388}
Metal['Zr'] = {'Mat': 91.22, 'Cs': 0.278}
Metal['Mo'] = {'Mat': 95.94, 'Cs': 0.251}
Metal['Ag'] = {'Mat': 107.86, 'Cs': 0.235}
Metal['Sb'] = {'Mat': 121.76, 'Cs': 0.207}
Metal['Ta'] = {'Mat': 180.95, 'Cs': 0.14}
Metal['W'] = {'Mat': 183.84, 'Cs': 0.132}
Metal['Au'] = {'Mat': 196.97, 'Cs': 0.129}
Metal['Pb'] = {'Mat': 207.2, 'Cs': 0.13}
Metal['Bi'] = {'Mat': 208.99, 'Cs': 0.122}
```

```python
from sklearn.linear_model import LinearRegression

Cs = []
Mat = []
for key in Metal.keys():
  Cs.append(Metal[key]['Cs'])
  Mat.append(1/Metal[key]['Mat'])

x = np.array(Cs).reshape((-1,1))
y = np.array(Mat)
model = LinearRegression()
model.fit(x,y)
print(f"the slope of the best line: {model.coef_}")

x_new = np.linspace(0,4,100).reshape((-1,1))
y_new = model.predict(x_new)

x_Be = np.array([[1.825]])
y_Be = model.predict(x_Be)
print(f"Predicted Atomic mass of Be: {1/y_Be[0]}")

fig = plt.figure(figsize=(8,3))
ax = fig.add_subplot(1,1,1)
ax.plot(Cs, Mat, 'ro')
ax.plot(x_new.reshape(1,-1)[0], y_new.reshape(1,-1)[0],'-b')

ax.set_xlabel('Cs (Jg$^{-1}$K$^{-1}$)', fontsize=15)
ax.set_ylabel('1/M$_{at}$', fontsize=15)

ax.grid(True)
plt.show()
```
