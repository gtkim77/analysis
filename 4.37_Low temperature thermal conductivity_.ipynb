```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def func(x, a, b):
  return a * x + b

T = np.array([1.29,1.44,1.59,1.79,2.04,2.45,2.82,3.21,3.59,4.13,4.7,5.36,5.93,6.96,8.02])
kappa = np.array([0.342,0.45,0.556,0.834,1.25,2.05,3.3,4.75,6.64,9.95,14.9,21.9,30,45.0,64.7])

lnT = np.log(T)
lnkappa = np.log(kappa)

popt, pcov = curve_fit(func, lnT, lnkappa)
print(popt)

plt.plot(lnT, lnkappa, 'ko')
plt.plot(lnT, func(lnT, *popt), 'b-')
plt.grid()
plt.show()
```
