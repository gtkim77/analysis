```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def func(x, a, b):
  return a * x + b

T = np.array([50,60,80,100,150,175,200,250,300,400,500,600,700,800])
kappa = np.array([5.9,4.7,3.1,2.25,1.3,1.1,0.95,0.73,0.6,0.44,0.338,0.269,0.219,0.193])


lnT = np.log(T)
lnkappa = np.log(kappa)

popt,perr = curve_fit(func, lnT, lnkappa)
print(popt)

plt.plot(lnT, lnkappa, 'ko')
plt.plot(lnT, func(lnT, *popt), 'b-')
plt.grid()
plt.xlabel('ln(T)')
plt.ylabel('ln($\kappa$)')
plt.show()
```
